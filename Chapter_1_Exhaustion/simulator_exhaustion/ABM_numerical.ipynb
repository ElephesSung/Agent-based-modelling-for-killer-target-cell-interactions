```python
ABM_results = {}
number_delta = 51
number_omega0 = 51
delta_list = np.round(np.linspace(0.25, 1.0, number_delta), 2)
omega0_list = np.round(np.linspace(0.5, 1.0, number_omega0), 2)

chi0 = 0.0
phi = 0.1
rho_min = 0
rho_max = 0.25

replicates = 25

for omega0 in tqdm(omega0_list, desc="Omega0 loop"):
    for delta in tqdm(delta_list, desc="Delta loop", leave=False):
        for r in tqdm(range(replicates), desc="Replicate loop", leave=False):
            ABM_results[(omega0, delta, r)] = KTSimulator(
                KILLER_STATE_INIT = omega0,
                TARGET_STATE_INIT = delta,
                MIN_STATE_DECAY_dt = rho_min,
                MAX_STATE_DECAY_dt = rho_max,
                DEATH_FAC_INIT = chi0, 
                RECOVERY_SPEED= phi,               
                )[0] # DataFrame
```
